# 10X Genomoics

```python
def rename_columns_with_binder(df):
    new_columns = {}
    for col in df.columns:
        if col.endswith('_binder'):
            # Split the column name by '_' and extract the second part
            new_col_name = col.split('_')[1]
            new_columns[col] = new_col_name
    df = df.rename(columns=new_columns)
    return df

def cleanup_10xGenomics(index):
    df = pd.read_csv(f'{path}10xGenomics\\vdj_v1_hs_aggregated_donor{index}_binarized_matrix.csv')
    df = df.rename(columns={'cell_clono_cdr3_aa': 'TCR'})
    df['TCR'] = df['TCR'].str.replace('TRA:', '').str.replace('TRB:', '')
    df = df.dropna()
    df['TCR'] = df['TCR'].str.split(';')
    df = df.explode('TCR')
    binding_columns = [col for col in df.columns if col.endswith('_binder')]
    epitops = [item.split('_')[1] for item in binding_columns]
    df = rename_columns_with_binder(df)

    negative_samples = pd.DataFrame()

    # Iterate over the binding columns and extract TCR-epitope pairs where binding is False
    for column in epitops:
        false_bindings = df[df[column] == False]
        
        tcr_sequences = false_bindings['TCR']
        epitope = column.split('_binder')[0]
        
        temp_df = pd.DataFrame({
            'TCR': tcr_sequences,
            'epitope': epitope,
            'binding': 0  # 0 for negative samples (non-binding)
        })
        
        negative_samples = pd.concat([negative_samples, temp_df])

    return negative_samples.drop_duplicates()
```

```python
df1 = cleanup_10xGenomics(1)
df2 = cleanup_10xGenomics(2)
df3 = cleanup_10xGenomics(3)
df4 = cleanup_10xGenomics(4)
df_10xgenomics = pd.concat([df1, df2, df3, df4], ignore_index=True, sort=False)
df_10xgenomics.to_csv(f'{path}Cleaned\\10xGenomics.csv', index=False)
```
